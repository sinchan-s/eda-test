### Final code:

```python
catg_list = joined_df['category_name'].unique()
joined_df2 = joined_df.sort_values(by='order_date', ascending=True)
days = joined_df2.order_date.unique()

for i in range(len(catg_list)):
    category = catg_list[i]
    #print(f'{category.upper()}\nDate Qty(Daily Penetration) Total')
    total_days = len(days)
    qty = [q for q in range(total_days)]
    for j in range(total_days):
        each_day = days[j]
        day_df = joined_df2[joined_df2.order_date == each_day]
        categories = day_df.value_counts('category_name')
        catg_sum = categories.sum()
        try:
            quantity = categories[categories.index == category].values[0]
            penetration = (100*int(quantity)/int(catg_sum))
            qty[j] = penetration
            #print(f'{each_day} {quantity}({penetration:.2f}%) {catg_sum}')
        except:
            qty[j] = 0
            #print(f'{each_day} 0(0%) {catg_sum}')
    print(f'\nAverage D-o-D Penetration of {category.upper()}: {sum(qty)/len(qty):.2f}%\n')
```
